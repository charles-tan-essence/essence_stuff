```python
import numpy as np
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from statsmodels.stats.proportion import proportions_ztest as ztest
from googleapiclient.discovery import build
import matplotlib.pyplot as plt
import seaborn as sns

from authentication.authenticator import Authenticator
from sheets.sheetmanager import SheetManager
```

```python
keys = 'credentials.json'
SCOPES = ['https://www.googleapis.com/auth/drive',
         'https://www.googleapis.com/auth/cloud-translation']

data_spreadsheetId = '1Yr-jmDKJixCcRRBK73yaWU-VtxZlcW5bnwm8wHqm538'
data_data_range = 'Data'
```

```python
authenticator = Authenticator(keys)
creds = authenticator.get_creds(SCOPES)
manager = SheetManager(creds)

resource = build('translate', 'v2', credentials=creds)
```

```python
def translate(string, source='ja'):
    string = str(string)
    if string != 'nan':
        return(resource.translations().list(q=string,
                                            source=source,
                                           target='en',
                                           format='text').execute()['translations'][0]['translatedText'])
    else:
        return('')
translate('わかりません')
```

```python
data_df = manager.get_values(spreadsheetId=data_spreadsheetId,
                        data_range=data_data_range)
```

```python
df = data_df.copy()

df['Q187_en'] = df.apply(lambda x: translate(x['Q187']), axis=1)
df['Q188_en'] = df.apply(lambda x: translate(x['Q188']), axis=1)
df
```

```python
df['Q187_tokens'] = df['Q187_en'].apply(word_tokenize)
df['Q188_tokens'] = df['Q188_en'].apply(word_tokenize)

list_of_words = df['Q187_tokens'].values

tokens = []

for list_ in list_of_words:
    for token in list_:
        tokens.append(token)

filtered_tokens = [token for token in tokens if (token.lower() not in stopwords.words('english') or len(token)>1)]    
        
lemmatized_tokens = []        
        
lemmatizer = WordNetLemmatizer()

for token in filtered_tokens:
    lemmatized_tokens.append(lemmatizer.lemmatize(token))
    
tokens_df = pd.DataFrame()

token_counts = {}

for token in lemmatized_tokens:
    if token not in token_counts.keys():
        token_counts[token] = 1
    else:
        token_counts[token] += 1
    
tokens_df = pd.DataFrame.from_dict(token_counts, orient='index')

tokens_df.sort_values(0, ascending=False)

#ax = sns.countplot(lemmatized_tokens)
```
