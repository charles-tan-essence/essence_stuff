2. For checkbox/multi-select questions, replace 0 with NaN, replace 1 (Unchecked) with 0, and replace 2 (checked) with 1. This will help when we use the weights.

3. Then multiply in the weights

```python
df = data_df.copy()

multi_select_questions = codebook_df[codebook_df['Question/Answer'].isin(['Not shown', 'Unchecked', 'Checked'])]['Variable'].unique().tolist()

df[multi_select_questions] = df[multi_select_questions].replace('', np.nan) - 1
df[multi_select_questions] = df[multi_select_questions].apply(pd.to_numeric)

multi_select_df = df[multi_select_questions].multiply(weights, axis=0).rename(columns=question_dic).add_suffix('_R')
multi_select_df
```

4. For likert scale, code by top 2 box etc.

```python
df = data_df.copy()

likert_dic = {
    'Q161': [0, 1],
    'Q176': [0, 1],
    'Q182': [0, 1],
    'Q183': [0, 1],
    'Q184': [0, 1],
    'Q185': [0, 1],
    'Q186': [0, 1], 
    'Q131': [2, 3],
}

likert_questions = list(likert_dic.keys())

likert_df = df[likert_questions]

def test(series):
    series = series.where(series.isin(likert_dic[series.name]), 'X')
    series = series.mask(series.isin(likert_dic[series.name]), 1)
    series[series=='X'] = 0
    series = pd.to_numeric(series)
    return(series)

likert_df = likert_df.apply(test, axis=0).mul(weights, axis=0).rename(columns=question_dic).add_suffix('_R')


likert_df.head()
```

Transform the data such that we have questions as rows, and also indexed by cell?

```python
table = recoded_df.copy()
weights_table = pd.DataFrame(weights).merge(table, left_index=True, right_index=True)    
bases = weights_table.groupby(table['Group']).sum()['Default Weights']
questions = [question for question in table.columns.tolist() if question != 'Group']

table = pd.pivot_table(table, values=questions, columns=['Group'], aggfunc='sum', dropna=False)

groups = table.columns
table = table.reset_index()
    
for group in groups:
    table[group+' Base'] = bases[group]
    table[group+' Desired %'] = table[group] / table[group+' Base']
    if group != control_name:
        table[group+' Abs Lift %'] = table[group+' Desired %'] - table[control_name+' Desired %']
        table[group+' p-value'] = table.apply(lambda x: ztest(
                                                            [x[group], x[control_name]],
                                                            [x[group+' Base'], x[control_name+' Base']]
                                                                )[1], axis=1)
        table[group+' Significance'] = 'No Lift'
        table[group+' Significance'].loc[table[group+' p-value'] < .2] = 'Directional Lift'
        table[group+' Significance'].loc[table[group+' p-value'] < .1] = 'Significant Lift'
        table[group+' Significance'].loc[table[group+' p-value'] < .05] = 'Strong Significant Lift'

table = table.set_index('index')        
        
suffixes = [' Base', ' Desired %', ' Abs Lift %', ' Significance'] # can include control_name+' p-value'

col_order = [control_name, control_name+' Base', control_name+' Desired %']
for col in groups:
    if col != control_name:
        col_order.append(col)
        for suffix in suffixes:
            col_order.append(col+suffix)
    
table = table[col_order]

table.to_csv('Overall Results.csv')
```
