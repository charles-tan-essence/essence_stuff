```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from authentication.authenticator import Authenticator
from sheets.sheetmanager import SheetManager
```

```python
keys = 'credentials.json'
SCOPES = ['https://www.googleapis.com/auth/drive.readonly']

spreadsheetId = '1WwND23IyNagMGgoH0desaARn7UYJWVGHnvF0Rar92iI'
data_ranges = ['18-24 Top channels',
               '25-34 Top channels',
               '35-44 Top channels',
               '45-54 Top channels']
```

```python
authenticator = Authenticator(keys)
creds = authenticator.get_creds(SCOPES)
manager = SheetManager(creds)
```

```python
data_df = None
for data_range in data_ranges:
    temp_df = manager.get_values(spreadsheetId=spreadsheetId,
                                data_range=data_range)
    temp_df['Age Cat'] = data_range
    if data_df is None:
        data_df = temp_df
    else:
        data_df = pd.concat([data_df, temp_df])
data_df    
```

```python
df = data_df.copy()

df.dtypes
df['Latest # Subs'] = pd.to_numeric(df['Latest # Subs'])

g = sns.barplot(x='Channel Vertical', y='Latest # Subs', hue='Age Cat', data=df)
type(g)
```
