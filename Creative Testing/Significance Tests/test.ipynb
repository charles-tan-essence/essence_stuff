```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import seaborn as sns

from authentication.authenticator import Authenticator
from sheets.sheetmanager import SheetManager
```

```python
data_url = 'https://docs.google.com/spreadsheets/d/1uYWz6oNDvUi3AuDoi8uxMmDJN92qPOGwdrsM7UJ7gRo/edit#gid=0'
data_sheet_name = 'Sheet1'

keys = 'credentials.json'
SCOPES = ['https://www.googleapis.com/auth/drive']

authenticator = Authenticator(keys)
creds = authenticator.get_creds(SCOPES)
manager = SheetManager(creds)

data_spreadsheetId = re.findall(r"/spreadsheets/d/([a-zA-Z0-9-_]+)",data_url)[0]
data_data_range = data_sheet_name

data_df = manager.get_values(spreadsheetId=data_spreadsheetId,
                        data_range=data_data_range)
```

```python
df = data_df.copy()

df['AVOC impressions'] = df['Impression'] * df['Valid and AVOC Rate']


df_grouped = df.groupby(['4C/Non-4C', 'Region', 'Ad Format', 'Age Group']).sum().reset_index()[['4C/Non-4C', 'Region', 'Ad Format', 'Age Group', 'Impression', 'AVOC impressions']]

df_grouped['AVOC rate'] = df_grouped['AVOC impressions']/df_grouped['Impression']

df_grouped

#sns.barplot(x='Age Group', y='AVOC rate', hue='Region', data=df_grouped)
```
